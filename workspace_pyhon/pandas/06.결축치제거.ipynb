```python
#엑셀 불러오기
import pandas
df = pandas.read_excel('test1.xlsx')
df
```

```python
#결측치 : 데이터 세트에서 누락된 값
#비어있는데이터, NaN 데이터 
#결측치를 제거하는 방법은 결측치를 삭제하는 방법, 특정데이터로 채우는 방법
```

```python
#실습을 위한 결측치 데이터 추가
import numpy

df['classRoom'] = numpy.nan
df
```

```python
#nan 데이터를 특정 데이터로 채우기
df.fillna('data',inplace=True)

#모든 컬럼의 nan데이터 수정
df
```

```python
#특정 컬럼의 NaN을 수정하기
df['classRoom'].fillna('1',inplace=True)

df
```

```python
#NaN 데이터 제거
#dropna의 매개변수

#axis = 'index' : NaN데이터를 기준으로 행을 삭제(defalut)
#axis = 'columns' : NaN데이터를 기준으로 열을 삭제
#how='any' : NaN 데이터가 하나라도 존재하면삭제(defalut)
#how='All' : 모든 데이터가 NaN일때 삭제
df.dropna(axis = 'index', how = 'any')

df
```
